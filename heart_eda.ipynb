```python
# Importing the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading the Dataset
df = pd.read_csv('heart_cleaned_data.csv')
display(df.head(1))
print(f"Number of Columns: {len(df.columns)}\nNumber of Rows: {len(df)}")
```

```python
# Checking the data Types of data
display(df.dtypes)
```

```python
# Disease Condition 
disease = df['Disease Condition'].value_counts()
plt.figure(figsize=(4, 4))
plt.pie(disease, labels=['No Disease', 'Disease'], autopct="%1.1f%%", colors=["#66b3ff", "#ff9999"])
plt.title("Disease Condition")
plt.show()
```

```python
# Disease Condition Distribution by Sex
disease_over_sex = df.groupby(['Sex', 'Disease Condition']).size().unstack(fill_value=0)
plt.figure(figsize=(6,5))
disease_over_sex.plot(kind='bar', stacked=False, ax=plt.gca(), cmap='viridis', color=['yellow', 'orange'], alpha=0.8)
plt.title('Disease Condition Distribution by Sex', fontsize=16)
plt.xlabel('Sex', fontsize=14)
plt.xticks(rotation=0)
plt.legend(title='Disease Condition')
plt.tight_layout()
plt.show()
```

```python
# Disease Condition Distribution by Age
sns.histplot(data=df, x='Age', hue='Disease Condition', color=['blue', 'red'], bins=20, kde=True)
plt.title('Disease Condition Distribution by Age', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
chest_pain_type = df['Chest Pain Type'].value_counts()
plt.pie(chest_pain_type, labels=['Asymptomatic', 'Non-anginal Pain', 'Atypical Angina', 'Typical Angina'], autopct="%1.1f%%")
plt.title('Chest Pain Type', fontsize=16)
plt.show()
```
